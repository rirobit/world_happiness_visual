```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
score_countries_page = requests.get("https://en.wikipedia.org/wiki/World_Happiness_Report#2020_report")
score_countries_soup = BeautifulSoup(score_countries_page.content, 'lxml')
```

```python
score_countries_table = score_countries_soup.find_all('table', {'class':'wikitable'})
score_countries_table[1]
```

```python
countries_score_df = pd.read_html(str(score_countries_table[1]))
countries_score_df
```

```python
countries_score_df = countries_score_df[0]
countries_score_df
```

```python
countries_score_df = countries_score_df.rename(columns={"Country or region":"Country"})
countries_score_df
```

```python
continent_page = requests.get("https://statisticstimes.com/geography/countries-by-continents.php")
continent_soup = BeautifulSoup(continent_page.content, 'lxml')
continent_soup
```

```python
continent_table = continent_soup.find('table', {'id':'table_id'})
continent_table
```

```python
continent_df = pd.read_html(str(continent_table))
continent_df = continent_df[0]
continent_df
```

```python
continent_df = continent_df.rename(columns={"Country or Area":"Country"})
continent_df = continent_df.drop(columns=['ISO-alpha3 Code', 'M49 Code', 'Region 1','Region 2'])
continent_df
```

```python
merge_df = pd.merge(countries_score_df, continent_df, how='inner', on='Country')
merge_df = merge_df.drop(columns="No")
merge_df.to_csv("final_result.csv")
merge_df
```

```python
score_df = pd.read_csv("final_result.csv", index_col=2)
score_df
```

```python
ax = score_df['Score'].plot(kind='bar', figsize=(30,10), title="Happiness Score For Each Country")
ax.set_ylabel("Happiness Score")
```

```python
plt.figure(figsize=(30,15))
plt.title("Histogram of Number of Countries Wihtin a Range of Happiness Score")
plt.xlabel("Happiness Score")
plt.ylabel("Number of Countries")
plt.hist(score_df["Score"], bins=9)
plt.show()
```

```python
median_happiness_score = merge_df["Score"].median()
countries_above_median = merge_df.loc[merge_df["Score"] > median_happiness_score]
print(countries_above_median["Country"].to_list())
print(len(countries_above_median), "countries are above median")
```

```python
heatmap_df = merge_df.drop(['Overall rank','Country','Continent'], axis=1)
ax = sns.heatmap(heatmap_df.corr(), annot=True, fmt='.2f', cmap='Greens')
```
